```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
x=np.outer(np.linspace(-0.2,0.2,50),np.ones(50))
y=x.copy().T
z=np.cos(x//2+y//4)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,z,cmap='tab20c',ec='k')
ax.set_title('surfaceplot 3d')
plt.show()
```
